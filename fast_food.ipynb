# Fast Food Nutrition Dataset

### Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import data

```python
ffood = pd.read_csv("fastfood.csv")
ffood.head()
```

```python
ffood.info()
```

### Capitalize column headers

```python
ffood.columns = ffood.columns.str.title()

ffood.tail()
```

### Search for NaN values in the Vit_C, Vit_A, and Calcium columns

```python
ffood[ffood["Vit_C"].isnull()] # found 210 NaN values
```

```python
ffood[ffood["Vit_A"].isnull()] # found 214 NaN values
```

```python
ffood[ffood["Calcium"].isnull()] # found 210 NaN values
```

### Fill the NaN values with zeros

```python
# Like the NaN values in these three columns represents more that the 40% of the data
# the recomendation is to fill this values with zeros because other information is not available
```

```python
ffood.fillna(0, inplace = True)
```

```python
ffood.info()
```

### Data Analysis

```python
ffood.head()
```

```python
# I want to know wich restaurants are in the dataset

ffood["Restaurant"].unique()
```

```python
# Resturant with the bigger number of item

ffood["Restaurant"].value_counts()
```

```python
# Percentage that represent the items per restaurant over the total

total = ffood["Item"].nunique()
total
```

```python
item = ffood["Restaurant"].value_counts()
percentage = (item*100)/total
percentage
```

```python
data = percentage.values
labels = percentage.index

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data, labels = labels ,colors = colors, autopct = '%.1f%%')
plt.title("Percentage of Item per Restaurant")
```

```python
# Top 3 Items with most calories

ffood.sort_values(by = "Calories", ascending = False).head(3)
```

```python
# Top 3 Items with most Cholesterol

ffood.sort_values(by = "Cholesterol", ascending = False).head(3)
```

```python
# Top 3 Items with most Sugar

ffood.sort_values(by = "Sugar", ascending = False).head(3)
```

```python
# Average value of Fiber and Total Carbs by Fast Food Restaurants

table = ffood.pivot_table(values = ["Total_Carb", "Fiber", "Protein"], index = "Restaurant", aggfunc = "mean") 

sort_table = table.sort_values(by = "Total_Carb", ascending = True)

sort_table
```

```python
# Plot the results

sort_table.plot(kind = "bar")
plt.xticks(rotation = 30)
plt.xlabel("Restaurants")
plt.ylabel("Mean")
plt.title("Average of Nutrients by Fast Food Restaurants")
```

```python
# Average of Total Fat, Sat Fat, and Trans Fat by Fast Food Restaurants

table1 = ffood.pivot_table(values = ["Total_Fat", "Sat_Fat", "Trans_Fat"], index = "Restaurant", aggfunc = "mean")

sort_table1 = table1.sort_values(by = "Total_Fat", ascending = True)

sort_table1
```

```python
# Plot the results

sort_table1.plot(kind = 'barh', stacked = True, color = ['red', 'skyblue', 'green'])
plt.xlabel("Mean")
plt.ylabel("Restaurant")
plt.title("Average of Fat type by Fast Food Restaurants")
```
